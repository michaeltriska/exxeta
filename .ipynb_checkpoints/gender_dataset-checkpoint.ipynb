```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import array
from numpy import argmax

df_train = pd.read_csv('etc/notebooks/data/train.csv', sep=',')
df_test = pd.read_csv('etc/notebooks/data/test.csv', sep=',')
print(df_test)
df = df_train.append(df_test)
print(df.head())
print(df.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

obj_df = df_train.copy()

obj_df["time_extract_daytime"] = obj_df["time_extract_daytime"].astype('category')
obj_df["time_extract_daytime"] = obj_df["time_extract_daytime"].cat.codes

obj_df['general_female_websitescore'] = \
    (obj_df['general_female_websitescore']-obj_df['general_female_websitescore'].mean())/obj_df['general_female_websitescore'].std()
obj_df['general_male_websitescore'] = \
    (obj_df['general_male_websitescore']-obj_df['general_male_websitescore'].mean())/obj_df['general_male_websitescore'].std()
obj_df['user_no_of_sitevisits'] = \
    (obj_df['user_no_of_sitevisits']-obj_df['user_no_of_sitevisits'].mean())/obj_df['user_no_of_sitevisits'].std()
obj_df['user_extract_avg_time_per_page'] = \
    (obj_df['user_extract_avg_time_per_page']-obj_df['user_extract_avg_time_per_page'].mean())/obj_df['user_extract_avg_time_per_page'].std()
obj_df['time_extract_weekend'] = \
    (obj_df['time_extract_weekend']-obj_df['time_extract_weekend'].mean())/obj_df['time_extract_weekend'].std()
obj_df['user_avg_no_of_sitevisits_per_day'] = \
    (obj_df['user_avg_no_of_sitevisits_per_day']-obj_df['user_avg_no_of_sitevisits_per_day'].mean())/obj_df['user_avg_no_of_sitevisits_per_day'].std()

obj_df[obj_df.isnull().any(axis=1)]
```

```python
X = obj_df[['general_female_websitescore',
            'general_male_websitescore',
            'user_no_of_sitevisits',
            'user_extract_avg_time_per_page',
            'time_extract_weekend',
            'user_avg_no_of_sitevisits_per_day']]

from sklearn.manifold import TSNE
n_sne = 7000
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)

print("training tsne..")
X_2d = tsne.fit_transform(X)
print("training end")
```

```python
df = pd.DataFrame(X_2d, columns=['x', 'y'])
categories = df_train['gender'].values
df['category'] = categories
print(df.groupby('category').count())

plt.figure(figsize=(15, 15))
colors = 'r', 'g'
colormap = {'f':'black', 'm':'green'}
for i in range(len(categories)):
    plt.scatter(df.iloc[i].x,
                df.iloc[i].y,
                c=colormap[df.iloc[i].category])
plt.show()
```

```python
from sklearn import svm
from sklearn.cross_validation import train_test_split
from sklearn.metrics import classification_report

y = categories
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, train_size = 0.8, random_state=2)

clf = svm.SVC()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(classification_report(y_test, predictions))
```
